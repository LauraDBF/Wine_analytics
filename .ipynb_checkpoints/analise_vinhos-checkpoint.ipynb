```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import folium
from geopy.geocoders import Nominatim
import numpy as np
import seaborn as sns
```

```python
dir_script = os.getcwd()

pasta_csv = os.path.join(dir_script, 'dados_vinhos_kaggle') #nome da pasta

dfs = []

#for de leitura dos arquivos
for arquivo in os.listdir(pasta_csv):
    if arquivo.endswith('.csv'):
        caminho_arquivo = os.path.join(pasta_csv, arquivo)
        
        try:
            df = pd.read_csv(caminho_arquivo) #leitura de arquivo
            dfs.append(df)
        except Exception as e:
            print(f"Erro ao ler o arquivo {arquivo}: {str(e)}") #caso o arquivo esteja quebrado

            #concatendo e gerando um df unico com todos os arquivos csv que não quebraram
df_wine = pd.concat(dfs, ignore_index=True)
```

```python
#visualizar o dataset de vinhos para entender melhor sua estrutura 
df_wine.head()
```

# Idéias de análises para esses dados de vinhos

- Agrupamento dos tipos do uvas
- Separar as fabricantes de vinhos e mostrar quais tipos de uvas cada fabricante possui
- Identificar cada um dos testadores dos vinhos, e quais são suas especialidades de uvas
- Separar quais são as uvas mais produzidas por cada um dos paises e também as provincias
- Media e mediana de pontos dos vinhos por preço, encontrando os vinhos com melhor custo beneficio por tipo de vinho

Geração de um dashboard interativo com todos esses paramêtros 

## Media dos pontos

```python
media_points = df_wine['points'].mean()
print(f"A média dos valores na coluna 'points' é: {media_points}")
```

## Grafico de pontos

```python
plt.figure(figsize=(8, 6))
plt.hist(df_wine['points'], bins=20, edgecolor='k', color ="skyblue")
plt.xlabel('Points')
plt.ylabel('Frequência')
plt.title('Distribuição dos Valores da Coluna Points')
plt.show()
```

## Média de preços

```python
media_price = df_wine['price'].mean()
print(f"A média dos valores na coluna 'price' é: {media_price}")
```

## Grafico de preços

```python
plt.figure(figsize=(8, 6))
plt.hist(df_wine['price'], bins=10, edgecolor='k', color="green")
plt.xlabel('Price')
plt.ylabel('Frequência')
plt.title('Distribuição dos Valores da Coluna Price')
plt.show()
```

```python
#contagem
price_counts = df_wine['price'].value_counts().reset_index()
price_counts.columns = ['price', 'frequencia']

#grafico de dispersao (scatter plot) usando o sns
plt.figure(figsize=(10, 6))
sns.scatterplot(x='price', y='frequencia', data=price_counts)
plt.axhline(y= media_price, color = "red", linestyle="--", label= f'Média: {media_price:.2f}')
plt.xlabel('Preço dos rôtulos')
plt.ylabel('Frequência')
plt.title('Distribuição Numérica entre Preços e Frequência')

#legenda dentro da imagem pra colocar a média dos preços
plt.annotate(f'Média: {media_price:.2f}', xy=(0.80, 0.13), xycoords='figure fraction', color='red', fontsize=12, ha='right', va='bottom')

plt.show()
```
